```python
# define the parameter grid
param_grid = {
    'n_components': [1,3,5,7],
    'doc_topic_prior': [0.1, 0.25, 0.5],
    'topic_word_prior': [0.01, 0.1, 1]
}

# create an LDA estimator
lda = LatentDirichletAllocation(max_iter=100)

# compute log-likelihood for each combination of hyperparameters
scores = {}
for n_components in param_grid['n_components']:
    for doc_topic_prior in param_grid['doc_topic_prior']:
        for topic_word_prior in param_grid['topic_word_prior']:
            lda.set_params(n_components=n_components, doc_topic_prior=doc_topic_prior,
                           topic_word_prior=topic_word_prior)
            lda.fit(bow_corpus)
            score = lda.score(bow_corpus)
            scores[(n_components, doc_topic_prior, topic_word_prior)] = score

# print the best hyperparameters and score
best_hyperparams = max(scores, key=scores.get)
best_score = scores[best_hyperparams]
print("Best hyperparameters:", best_hyperparams)
print("Best score:", best_score)
```

```python
from nltk.sentiment.vader import SentimentIntensityAnalyzer
analyzer = SentimentIntensityAnalyzer()
```

```python
def analyze_sentiment_vader_lexicon(review, threshold = 0.1, verbose = False):
    scores = analyzer.polarity_scores(review)
    binary_sentiment = 'positive' if scores['compound'] >= threshold else 'negative'
    if verbose:
        print('VADER Polarity (Binary):', binary_sentiment)
        print('VADER Score:', round(scores['compound'], 2))
    return binary_sentiment,scores['compound']
```

```python
for sentenceId in range(len(normalized_sentences)):
    print('\nReview text:\n')
    print(sentences[sentenceId],"\n")
    final_sentiment = analyze_sentiment_vader_lexicon(normalized_sentences[sentenceId],
                                                        threshold=0,
                                                        verbose=True)
    print('-'*60)
```

```python
sentiment_list = []

for sentenceId in range(len(normalized_sentences)):
    final_sentiment, sentiment_score = analyze_sentiment_vader_lexicon(normalized_sentences[sentenceId],
                                                                       threshold=0)
    sentiment_list.append(sentiment_score)
df = pd.DataFrame(sentiment_list)
topiced_corpus_df = pd.concat([df, topiced_corpus_df], axis=1)
```

```python
plt.figure(figsize=(10, 6))
sns.histplot(df['sentiment_score'], bins=30, kde=True)
plt.title('Distribution of Sentiment Scores')
plt.xlabel('Sentiment Score')
plt.ylabel('Frequency')
plt.show()
```

```python
topic_cols = ['Topic_0', 'Topic_1', 'Topic_2', 'Topic_3', 'Topic_4', 'Topic_5']
avg_probs = df[topic_cols].mean()

plt.figure(figsize=(10, 6))
sns.barplot(x=avg_probs.index, y=avg_probs.values)
plt.title('Average Probability for Each Topic')
plt.ylabel('Average Probability')
plt.show()
```

```python
for topic in topic_cols:
    plt.figure(figsize=(10, 6))
    sns.scatterplot(x=df['sentiment_score'], y=df[topic])
    plt.title(f'Sentiment Score vs. Probability for {topic}')
    plt.xlabel('Sentiment Score')
    plt.ylabel(f'Probability of {topic}')
    plt.show()
```
